```python
X_test = test_data[feautre]
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_val, y_train1, y_val = train_test_split(X, y, random_state = 0)

#Fit the data with a model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=20, random_state=0)

model.fit(X_train1, y_train1)
predict_val = model.predict(X_val)
```

```python

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_val, predict_val))
print('Mean Squared Error:', metrics.mean_squared_error(y_val, predict_val))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_val, predict_val)))
```
